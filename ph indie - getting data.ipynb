```python
def related_artists(artists):
    """
    Submit an array of artist ids
    Returns dataframe of related artists
    """
    more = pd.DataFrame()
    
    for artist in artists: 
        more_artists = requests.get("https://api.spotify.com/v1/artists/{}/related-artists".format(artist),headers= header)
        json_moreartists = json.loads(more_artists.text)
        num = len(json_moreartists['artists'])
        
        for artist in range(num):
            more = more.append(json_moreartists['artists'][artist], ignore_index=True)
            more['followers'] = json_moreartists['artists'][artist]['total']

    
    return(more)
```

```python
def artists_albums(artist_id):
    """
    Submit id of artist
    Returns dataframe of all the artists albums
    """
    
    albums = pd.DataFrame()
    response_albums = requests.get("https://api.spotify.com/v1/artists/{}/albums?include_groups=album,single".format(artist_id), 
                                   headers= header)
    json_albums = json.loads(response_albums.text)
    for idx in range(len(json_albums['items'])):
        json_albums['items'][idx]['artist_id'] = artist_id
        json_albums['items'][idx]['external_urls'] = json_albums['items'][idx]['external_urls']['spotify']
        if len(json_albums['items'][idx]['images']) > 1:
            json_albums['items'][idx]['image_640'] = json_albums['items'][idx]['images'][0]['url']
            json_albums['items'][idx]['image_300'] = json_albums['items'][idx]['images'][1]['url']
            json_albums['items'][idx]['image_64'] = json_albums['items'][idx]['images'][2]['url']
        albums = albums.append(json_albums['items'][idx], ignore_index=True).drop(['available_markets','images'], axis=1)
    return(albums)
```

```python
def albums_tracks(album_id):
    """
    Submit id of album
    Returns dataframe of all the tracks in album
    """
    
    tracks = pd.DataFrame()
    response_tracks = requests.get("https://api.spotify.com/v1/albums/{}/tracks".format(album_id), 
                                   headers= header)
    json_tracks = json.loads(response_tracks.text)
    for idx in range(len(json_tracks['items'])):
        tracks = tracks.append(json_tracks['items'][idx], ignore_index=True).drop(['available_markets'], axis=1)
        tracks['album_id'] = album_id

    return(tracks)
```

```python
def audio_features(track_id):
    """
    Submit list of track ids (max 100 tracks)
    Returns dataframe of features per track
    """
    
    audio_feature = pd.DataFrame()
    url_args = ",".join([track for track in track_id])
    response_feature = requests.get("https://api.spotify.com/v1/audio-features?ids={}".format(url_args), 
                                   headers= header)
    json_tracks = json.loads(response_feature.text)
    for idx in range(len(json_tracks['audio_features'])):
        audio_feature = audio_feature.append(json_tracks['audio_features'][idx], ignore_index=True)
    return(audio_feature)
```

```python
def artists_top_tracks(artist_id):
    """
    Submit id of artist
    Returns dataframe of artists top tracks
    """
    
    top_tracks = pd.DataFrame()
    response_top_tracks = requests.get("https://api.spotify.com/v1/artists/{}/top-tracks?country=PH".format(artist_id), headers= header)
    json_top_tracks = json.loads(response_top_tracks.text)
    for idx in range(len(json_top_tracks['tracks'])):
        json_top_tracks['tracks'][idx]['artist_id'] = artist_id
        json_top_tracks['tracks'][idx]['external_url'] = json_top_tracks['tracks'][idx]['external_urls']['spotify']
        if len(json_top_tracks['tracks'][idx]['album']['images']) > 1:
            json_top_tracks['tracks'][idx]['image_640'] = json_top_tracks['tracks'][idx]['album']['images'][0]['url']
            json_top_tracks['tracks'][idx]['image_300'] = json_top_tracks['tracks'][idx]['album']['images'][1]['url']
            json_top_tracks['tracks'][idx]['image_64'] = json_top_tracks['tracks'][idx]['album']['images'][2]['url']
        top_tracks = top_tracks.append(json_top_tracks['tracks'][idx], ignore_index=True).drop(['external_ids','type','external_urls'], axis=1)
    return(top_tracks)
```
